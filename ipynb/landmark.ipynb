**Landmarks Identification**

```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import dlib
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/V2F/driveage_gender_Race.csv")
df.head()
```

```python
path = df['URL'][0]
img = cv2.imread(path)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
def func(path):
  img = cv2.imread(path)
  detector = dlib.get_frontal_face_detector()
  predictor = dlib.shape_predictor("/content/drive/MyDrive/V2F/shape_predictor_68_face_landmarks.dat")
  faces = detector(img)
  for face in faces:
    landmarks = predictor(img,face)
  t = []
  for i in range(68):
    x = landmarks.part(i).x
    y = landmarks.part(i).y
    # cv2.circle(img,(x,y),3,(255,0,0),-1)
    t.append((x,y))
  # plt.imshow(img)
  return t
```

```python
d = pd.DataFrame()
d['path'] = df['URL']
l = len(df['URL'])
for i in range(68):
  d["o" + str(i+1)] = [(0,0) for i in range(l)]
d.head()
```

```python
func(d["path"][1])#landmarks for the sample image
```

Running for all the image

```python
nc = []
for i in range(l):
  try:
    t = func(d["path"][i])
    for j in range(68):
      d["o"+str(j+1)][i] = t[j]
  except:
    nc.append(i) 
    print("nc = ",len(nc))
  print(i)
```

images for which landmarks are not detected

```python
dataDict = {
    'Not Correct': nc
}
ncDf = pd.DataFrame(dataDict)
ncDf.to_csv('nc.csv')
```
