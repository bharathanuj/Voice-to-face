# **First Unzipping the Zip file which contains videos**

```python
from zipfile import ZipFile
  
  
# opening the zip file in READ mode
with ZipFile('/content/drive/My Drive/Video_test_mp4.zip', 'r') as zip:
    # printing all the contents of the zip file
    zip.printdir()
  
    # extracting all the files
    print('Extracting all the files now...')
    zip.extractall()
    print('Done!')
```

```python
listOfiles = zip.namelist()
```

# **Preprocessing Url to get perfect video name **

```python
for elem in listOfiles:
  dp = elem.split('/')
  if len(dp)==4:
    if dp[3]!= "":
      print(elem)
```

```python
import cv2
import os
```

```python
if not os.path.exists('PicsMp4frames'):
  os.makedirs('PicsMp4frames')
```

```python
from google.colab.patches import cv2_imshow
```

# **Extracting Image Frame from Video and storing it in a new folder**

```python
for elem in listOfiles:
  dp = elem.split('/')
  if len(dp)==4:
    if dp[3]!= "":
      key = dp[3].split('.')
      vid = cv2.VideoCapture(elem)
      frame = 0
      while True:
        success, fr = vid.read()
  
        frame += 1

        if frame>5:
          cv2.imwrite('./PicsMp4frames/frame' + dp[1]+dp[2]+key[0]+'.jpg',fr)
          break

      vid.release()
      cv2.destroyAllWindows()
```
