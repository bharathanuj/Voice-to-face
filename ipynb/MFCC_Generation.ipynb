trim_audio is used to trim the audio file by extracting last 5 seconds of video

```python
def trim_audio(f):
  sound = AudioSegment.from_file(f)
  try :
    temp = sound[-5*1000:]
    temp.export("/content/drive/My Drive/Audio/test_audio/t.mp4", format="mp4")
  except:
    sound.export("/content/drive/My Drive/Audio/test_audio/t.mp4", format="mp4")
```

audio_mfcc is used to extract mfcc from audio

```python
def audio_mfcc(f):
  y, sr = librosa.load("/content/drive/My Drive/Audio/test_audio/t.mp4", sr=22050)
  hop_length = 512
  MFFC = librosa.feature.mfcc(y,hop_length = hop_length, n_mfcc=13)
  img = librosa.display.specshow(MFFC,sr = sr,hop_length = hop_length)
  plt.savefig(f[:-4]+".png", bbox_inches="tight", pad_inches=0.0)
```
