# **Here we are using VGG16 Model for gender detection**

```python
from keras.applications.vgg16 import VGG16
#Import the datagenerator to augment images'
from keras.preprocessing.image import ImageDataGenerator
vgg=VGG16(include_top=False,weights='imagenet',input_shape=(200,200,3))
```

```python
for layer in vgg.layers:
  layer.trainable = False
```

```python
from keras.layers import Flatten,Dense,BatchNormalization,Activation,Dropout
from keras.models import Model
x1 = Flatten()(vgg.output)
#class1 = Dense(1024, activation='relu')(x)
batch = BatchNormalization()(x1)
prediction = Dense(2, activation='softmax')(batch)
model = Model(inputs=vgg.input, outputs=prediction)
model.summary()
```

```python
model.summary()
```

```python
from tensorflow.keras import optimizers
adam = optimizers.Adam()
model.compile(loss='sparse_categorical_crossentropy',
              optimizer=adam,
              metrics=['accuracy'])
```

```python
from datetime import datetime
from keras.callbacks import ModelCheckpoint, EarlyStopping


checkpoint = ModelCheckpoint(filepath='mymodel1.h5', 
                               verbose=2, save_best_only=True)
callback = EarlyStopping(patience = 3,monitor = 'loss')

start = datetime.now()


evl = model.fit(x_train,y_train,batch_size = 150,epochs = 10,verbose = 1,callbacks = [callback, checkpoint],validation_split = 0.2) 


duration = datetime.now() - start
print("Training completed in time: ", duration)
```
